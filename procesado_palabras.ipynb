```python
import numpy as np
import csv
import re
```

```python
text = ''
with open('mensajes_slack.csv','r') as f:
    data_file = csv.reader(f)
    
    for row in data_file:
        text += row[2] + ''
```

```python
# Separar emoticonos juntos:
text = text.replace("::",": :")

# Dejar un espacio delante de las letras mayúsculas:
text = re.sub('([A-Z]{1})', r' \1',text)

# Separar emoticonos del resto del texto:
text = re.sub('(:*:)',r' \1', text)

# Texto a minúscula:
text = text.lower()

# Separar el texto en palabras individuales
words = re.split(r"[\s\.,\?, \!, \¿, \¡, \-, \(, \), \<, \>]+",text)


# Quitar duplicados de palabras
unique_words = sorted(set(words))
print(len(unique_words))
```
